```python

#import pandas and import the data file 
#check the data file by access to a column

import pandas as pd
data = pd.read_csv("student.csv")
data['result']
```

```python
# declare the encoding column to variable
result_category = data["result"]
```

```python
# using method to give numerical values to the categories
### in this manner the model could be identified there are 
# some relationships between the categories 

from sklearn.preprocessing import LabelEncoder
obj = LabelEncoder()
result = obj.fit_transform(result_category)
```

```python
result
```

```python
# One-hot encoding
# get the values and catogeries are seperate classes

from sklearn.preprocessing import LabelBinarizer
obj = LabelBinarizer()
result = obj.fit_transform(result_category)
```
